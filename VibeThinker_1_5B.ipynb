```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="WeiboAI/VibeThinker-1.5B")
messages = [
    {"role": "user", "content": "Who are you?"},
]
pipe(messages)
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import TextStreamer

tokenizer = AutoTokenizer.from_pretrained("WeiboAI/VibeThinker-1.5B")
streamer = TextStreamer(tokenizer)
model = AutoModelForCausalLM.from_pretrained("WeiboAI/VibeThinker-1.5B")
messages = [
    {"role": "user", "content": "Who are you?"},
	  {"role": "user", "content": "Djkstra: sample code in classic C++ with sample graph of 10 nodes."},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=1000,   streamer=streamer) #40
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```
